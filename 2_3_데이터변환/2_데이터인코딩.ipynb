```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder

obj = {
    "Class": ['A', 'B', 'A', 'C', 'E', 'D', 'D', 'A', 'F']
}

df = pd.DataFrame(obj)

encoder = LabelEncoder()
encoder.fit(df.Class) # encoder.fit(df["Class"])
labels = encoder.transform(df.Class)

df["to_수치형데이터"] = labels

df
```

```python
obj2 = {
    "과일": ['사과', '포도', '수박', '딸기', '배', '사과', '사과', '수박']
}

df2 = pd.DataFrame(obj2)

encoder = LabelEncoder()
encoder.fit(df2)
labels2 = encoder.transform(df2)

df2["to_수치데이터"] = labels2 
df2

```

```python
import numpy as np 

conditionList = [
    (df["Class"] == "A"),
    (df["Class"] == "B"),
    (df["Class"] == "C"),
    (df["Class"] == "D"),
    (df["Class"] == "E"),
    (df["Class"] == "F"),
]

choiceList = list(range(6))

np_labels = np.select(condlist=conditionList, choicelist=choiceList)

df["Numpy_수치형데이터"] = np_labels
df
```
