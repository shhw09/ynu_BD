```python
import pandas as pd

obj1 = {
    "student_1d" : ['s1', 's2', 's3', 's4', 's5', 's6'],
    'score' : [55, 90, 85, 71, 63, 99],
}

df1 = pd.DataFrame(obj1)
df1

obj2 = {
    "student_1d" : ['t1', 't2', 't3', 't4', 't5', 't6'],
    'score' : [65, 99, 87, 75, 57, 88],
}

df2 = pd.DataFrame(obj2)
df2
```

```python
pd.concat([df1, df2], axis=0) # 0 방향이 세로방향
```

```python
pd.concat([df1, df2], axis=1)
```

```python
# 데이터 병합

obj1 = {
    "student_id" : ['s2', 's4', 's5', 's6'],
    'score' : [85, 71, 63, 99],
}

df1 = pd.DataFrame(obj1)

obj2 = {
    "student_id" : ['s1', 's2', 's3', 's4'],
    'score' : [35, 99, 87, 75],
}

df2 = pd.DataFrame(obj2)
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2, on="student_id")
```

```python
df1.merge(df2, how="outer", on="student_id")
```

```python
df1.merge(df2, how="left", on="student_id")
```

```python
df1.merge(df2, how="right", on="student_id")
```
