```python
import pandas as pd
```

```python
obj = {
    "Name": ["Olivia", "Lucas", "Sofia", "Zoe", "Ava", "Elliot"],
    "Age" : [22, 32, 27, 18, 38, 18],
    "Score" : [100, 95, 60, 77, 83, 84]
}
df = pd.DataFrame(obj)
df
```

```python
df.sort_values('Score', inplace=True)
```

```python
df
```

```python
df.sort_values('Score', ascending=False) # True는 오름차순/False는 내림차순
```

```python
df
```

```python
# sort_values()

df.sort_values(["Age", "Score"], ascending=[True, False])
```

```python
# sort_index()

rowmn = ["Olivia", "Lucas", "Sofia", "Zoe", "Ava", "Elliot"]

obj = {
    "Age" : [22, 32, 27, 18, 38, 18],
    "Score" : [100, 95, 60, 77, 83, 84]
}

df = pd.DataFrame(obj, index=rowmn)
df
```

```python
df.sort_index(ascending=False)
```

```python
df.sort_index(axis=1)
```

```python
# rank()

obj = {
    "Score" : [100, 95, 60, 77, 83, 85, 90, 80, 88, 75, 90, 54, 48, 84, 73],
    "Age" : [22, 22, 27, 18, 18, 19, 24, 26, 30, 27, 25, 21, 20, 17, 20],
}

df = pd.DataFrame(obj)
df
```

```python
# (실행안됨,,)df.sort_values("Score", ascending=False. inplace=True)
df.head()
```

```python
df["rank_avg1"] = df["Score"].rank()
```

```python
df.head()
```

```python
df["rank_avg2"] = df["Score"].rank(ascending=False)
df.head()
```

```python
df["rank_min"] = df["Score"].rank(method="min", ascending=False)
df.head
```

```python
df["rank_max"] = df["Score"].rank(method="max", ascending=False)
df.head()
```
